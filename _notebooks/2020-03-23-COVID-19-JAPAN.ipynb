## 東京経済　新型コロナウイルス国内感染の状況
https://github.com/kaz-ogiwara/covid19/blob/master/data/individuals.csv

```python
import pandas as pd
import altair as alt
import requests
import io

URL = "https://raw.githubusercontent.com/kaz-ogiwara/covid19/master/data/individuals.csv"

r = requests.get(URL)
df = pd.read_csv(io.BytesIO(r.content), sep=",")
df = df.drop(0).replace('代', '', regex=True)
df['date'] = df['確定年'].astype(str) + '-' + df['確定月'].astype(str) + '-' + df['確定日'].astype(str)
df
```

```python
alt.Chart(df).mark_line().encode(
    alt.X("date"),
    y='count()',
)
```

```python
alt.Chart(df).mark_line().encode(
    alt.X("date"),
    y='count()',
    color='居住地1'
)
```

```python
df2 = df[['居住地1']]
df2['date'] = pd.to_datetime(df['date'])
df2 = df2.set_index('date')
df2 = df2.resample('1D').count()
df2['date'] = df2.index
df2

```

```python
alt.Chart(df2).mark_bar().encode(
    x='date:T',
    y='居住地1:Q',
)
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X("年代", bin=True),
    y='count()',
)
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X("性別"),
    y='count()',
)
```
