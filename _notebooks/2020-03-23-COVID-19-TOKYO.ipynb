## 東京都 新型コロナウイルス陽性患者発表詳細 

https://catalog.data.metro.tokyo.lg.jp/dataset/t000010d0000000068


```python
import pandas as pd
import altair as alt
import requests
import io

URL = "https://stopcovid19.metro.tokyo.lg.jp/data/130001_tokyo_covid19_patients.csv"

r = requests.get(URL)
df = pd.read_csv(io.BytesIO(r.content), sep=",")
df = df.drop(0).replace('代', '', regex=True)
df
```

```python
alt.Chart(df).mark_line().encode(
    alt.X("公表_年月日"),
    y='count()',
)
```

```python
alt.Chart(df).mark_line().encode(
    alt.X("公表_年月日"),
    y='count()',
    color='患者_年代'
)
```

```python
df2 = df[['No']]
df2['date'] = pd.to_datetime(df['公表_年月日'])
df2 = df2.set_index('date')
df2 = df2.resample('1D').count()
df2['date'] = df2.index
df2
```

```python
alt.Chart(df2).mark_bar().encode(
    x='date:T',
    y='No:Q',
)
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X("患者_年代", bin=True),
    y='count()',
)
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X("患者_性別"),
    y='count()',
)
```
